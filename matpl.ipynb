```python
x=np.arange(0,np.pi*2,0.1)
y=np.sin(x)
z=np.tan(x)
t=np.cos(x)
fig=plt.figure()
axis=fig.add_axes([0,0,1,1])
axis.plot(x,y, color='black')
axis.plot(x,z, color='lightblue', marker='*', label='sine',linewidth=0.5)
axis.plot(x,t, color='blue', linestyle="-.", label='cosine') 
axis.legend(loc=1)#loc='upper right' we can also use label here
#axis.plot(x,x)
```

```python
x=np.arange(0,np.pi*2,0.1)
y=np.sin(x)
#z=np.tan(x)
t=np.cos(x)
fig=plt.figure()
axis=fig.add_axes([0,0,1,1])
axis.plot(x,y, color='black')
#axis.plot(x,z, color='lightblue', marker='*', label='sine')
axis.plot(x,t, color='blue', linestyle="-.", label='cosine') 
axis.legend(loc=1)#loc='upper right' we can also use label here
axis.plot(x,x)
```

# subplot

```python
fig, ax=plt.subplots(2,3)
plt.tight_layout
ax[0][0].plot(x,y)
ax[0][0].set_title('sine')
ax[0][1].plot(x,z)
ax[0][1].set_title('cos')
ax[0][2].plot(x,t)
ax[0][1].plot(t,y)
ax[1][0].plot(t,z)
ax[1][1].plot(z,y)
ax[1][2].plot(z,t)
ax[1][2].set_xlabel('angle')
```
